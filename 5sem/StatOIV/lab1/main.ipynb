Chennai House Price

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("chennai_house_price.csv")
```

```python
dataset
```

```python
important_columns = ["price", "area", "status", "bhk", "bathroom", "age"]
filtered_ds = dataset[important_columns].copy()
filtered_ds
```

```python
filtered_ds["status"] = filtered_ds["status"].replace({"Ready to move": 1, "Under Construction": 0}) # Делаем статусы целочисленными
filtered_ds.dropna(inplace=True) # Удалили строки со значением Nan
filtered_ds.reset_index(drop=True, inplace=True) # Пронумеровал id заново
filtered_ds.rename(columns={"bhk": "bedrooms"}, inplace=True)
filtered_ds
```

```python
filtered_ds.to_csv("filtered_chennai_house_price.csv")
```
